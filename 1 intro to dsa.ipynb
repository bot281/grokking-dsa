#### data structures are the foundational building blocks of cs and software development
first begin by understanding big O
* takes little practice and effort
* notation using math to describe performance of algorithm
* also describes its complexity
* most specifically, we want to know how long the algorithm takes to run as input size grows (scalability)
* allows analysis and efficiency comparison --> informed decisions on what to code
* provides rough estimate of how long an algorithm takes to run
* also estimates how much memory the program will use
* lower time complexity means faster (in most cases)
* space complexity measures memory based on size of input --> lower means less memory

#### time complexity
* O(1) constant time - run time independent of input size
* O(n) linear time - run time scales linearly with input size
* O(n<sup>2</sup>) quadratic time - run time proportional to square of input size
* O(logn) logarithmic time - run time increases log with input size
* O(2<sup>n</sup>) exponential time - run time increases exponentially in input size


* big O basically describes the upperbound of an algorithm's run time
